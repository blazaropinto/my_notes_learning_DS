```python
# Imports
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load dataset
eu_apples_prod = pd.read_csv('../datasets/clean/apples_eu.csv', index_col=0, header=0)
display(eu_apples_prod.head())
```

```python
# Set default size for figures
plt.rcParams['figure.figsize'] = (16, 5)

# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots()

# Plot apple production in Austria against the year
ax.plot(eu_apples_prod.index, eu_apples_prod["Austria"]/1000, 
        color= 'b', marker = 'o', linestyle= '--', label='Austria')

# Plot apple production in the UK against the year
ax.plot(eu_apples_prod.index, eu_apples_prod["United Kingdom of Great Britain and Northern Ireland"]/1000, 
        color= 'r', marker= 'v', linestyle ='-.', label='UK')

# Plot apple production in the UK against the year
ax.plot(eu_apples_prod.index, eu_apples_prod["Spain"]/1000, 
        color= 'k', marker= '*', linestyle='None', label='Spain')

# Set the x-axis label
ax.set_xlabel('Years')

# Set the y-axis label
ax.set_ylabel('Production of apples (thousands of tonnes)')

#Adding a title
ax.set_title('Production of apples in Austria, UK and Spain')

# Display the legend
ax.legend()

# Call the show function to display the resulting plot
plt.show()
```

### ZOOM

```python
# Use plt.subplots to create fig and ax
fig , ax = plt.subplots()

# Create variable seventies with data from "1970-01-01" to "1979-12-31"
seventies = eu_apples_prod.loc[1970 : 1979]

# Add the time-series for data from Albania from seventies to the plot
ax.plot(seventies.index, seventies["Albania"]/1000, color='g')

# Set the axis labels
ax.set_xlabel('Years')
ax.set_ylabel('Production of apples (thousands of tonnes)')

# Show the figure
plt.show()
```

### Secondary y axis

```python
# Initalize a Figure and Axes
fig, ax = plt.subplots()

# Plot the CO2 variable in blue
ax.plot(eu_apples_prod.index, eu_apples_prod["Albania"]/1000, 
        color='g', label='Albania', linestyle= ':')

# Create a twin Axes that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature in red
ax2.plot(eu_apples_prod.index, eu_apples_prod["Austria"]/1000, 
         label='Austria', color= 'b', linestyle= '--')

# Set the axis labels
ax.set_xlabel('Years')
ax.set_ylabel('Production of apples (thousands of tonnes)')

# Display the legend
ax.legend()
ax2.legend()

plt.show()
```

### Create funtion to to a customized plot

```python
# Define a function called plot_timeseries
def plot_prod_over_years(axes, x, y, color, xlabel, ylabel):

  # Plot the inputs x,y in the provided color
  axes.plot(x, y, color=color)

  # Set the x-axis label
  axes.set_xlabel(xlabel)

  # Set the y-axis label
  axes.set_ylabel(ylabel, color=color)

  # Set the colors tick params for y-axis
  axes.tick_params('y', colors=color)

fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_prod_over_years(ax, eu_apples_prod.index, eu_apples_prod['Sweden'],
                "purple", "Time (years)" , "Production of apples (thousands of tonnes)")

# Annotate point with relative temperature >1 degree
ax.annotate("What happened here?",xy=(1990, 165000), xytext=(2007,160000), fontsize=15,
             arrowprops={'arrowstyle':'->','color':'brown'})

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_prod_over_years(ax2, eu_apples_prod.index, eu_apples_prod['Belgium-Luxembourg'], 
                "red", "Time (years)", "Production of apples (thousands of tonnes)")

ax2.annotate("And here?",xy=(2000, 520000), xytext=(2007,300000), fontsize=15,
             arrowprops={'arrowstyle':'->','color':'gray'})

plt.show()
```

#### Sharing the y-axis range

```python
fig, ax = plt.subplots(2, 1, sharey=True)

ax[0].plot(eu_apples_prod.index, eu_apples_prod["Germany"]/1000, 
         label='Germany', color= 'b')
ax[0].legend()

ax[1].plot(eu_apples_prod.index, eu_apples_prod["Poland"]/1000, 
         label='Poland', color= 'c')
ax[1].legend()

# Set the axis labels
ax[1].set_xlabel('Years')
ax[0].set_ylabel('Production of apples (thousands of tonnes)')

plt.show()
```
