There are a number of pre-defined styles provided by Matplotlib. 

```python
# Imports
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# Load dataset
eu_apples_prod = pd.read_csv('../datasets/clean/apples_eu.csv', 
                             index_col=0, header=0)
display(eu_apples_prod.head())
```

```python
# def function to create example plot 
def create_example_plot():
    fig, ax = plt.subplots()

    # Plot apple production against the year
    ax.plot(eu_apples_prod.index, eu_apples_prod["Austria"]/1000, 
              label='Austria')
    ax.plot(eu_apples_prod.index, eu_apples_prod["United Kingdom of Great Britain and Northern Ireland"]/1000, 
              label='UK')

    # Set the labels
    ax.set_xlabel('Years')
    ax.set_ylabel('Production of apples (thousands of tonnes)')

    #Adding a title and display the legend
    ax.set_title('Production of apples in Austria and UK')
    ax.legend()

    # Call the show function to display the resulting plot
    plt.show()
    
    
# create plot with default parameters
create_example_plot()
```

```python
for style in plt.style.available:
    print(style)
    plt.style.use('default')
    plt.style.use(style)
    create_example_plot()
```

```python
plt.style.use('ggplot')
create_example_plot()
```

```python
plt.style.use('default')
```

You can create custom styles and use them by calling `style.use` with the path or URL to the style sheet

In the style sheet `.mplstyle` set default [parameters](https://matplotlib.org/stable/api/matplotlib_configuration_api.html#matplotlib.RcParams)

See [this link](https://matplotlib.org/stable/gallery/color/named_colors.html) for a list of the colour names

```python
plt.style.use('./mystyle.mplstyle')
create_example_plot()
```

The default rc settings can also be changed directly in a python script or interactively from the python shell. 
All of the rc settings are stored in a dictionary-like variable called matplotlib.rcParams, which can be modified directly using the same parameters

to use an image as the background for a plot,use the functions `imread()` and `imshow()`

```python
image = plt.imread('image.jpg')

fig, ax = plt.subplots(figsize=(15,8))
ax.imshow(image, aspect='auto', extent=[1955,2024,100,700])
ax.plot(eu_apples_prod.index, 
        eu_apples_prod["United Kingdom of Great Britain and Northern Ireland"]/1000, 
          label='UK', linestyle=':', linewidth=8)

ax.set_xlabel('Years')
ax.set_ylabel('Production of apples (thousands of tonnes)')
ax.set_title('Production of apples in the United Kingdom')
ax.legend()
plt.show()
```

To work with a colour palette, the arguments `c` (which needs to be an array of floats) and `cmap` (the [colormap from matplotlib](https://matplotlib.org/stable/tutorials/colors/colormaps.html) 

must be used instead of `color`

```python
plt.style.use('default')
fig, ax = plt.subplots(figsize=(15,8))
ax.scatter(eu_apples_prod.index, 
        eu_apples_prod["Austria"]/1000, 
          label='Austria',c=eu_apples_prod["Austria"]/1000, cmap='viridis')

ax.set_xlabel('Years')
ax.set_ylabel('Production of apples (thousands of tonnes)')
ax.set_title('Production of apples in Austria')
ax.legend()
plt.show()
```
