```python
# Imports
import matplotlib.pyplot as plt
import pandas as pd

# Set default size for figures
plt.rcParams['figure.figsize'] = (16, 5)
```

```python
df = pd.read_csv('../datasets/clean/simpledf_applevalues.csv', index_col=0)
df.head()
```

### Boxplots

```python
fig, ax = plt.subplots()

ax.boxplot([df.Br, df.Pt, df.Sp, df.UK])

ax.set_ylabel("Production of apples (tonnes)")
ax.set_xticklabels(df.columns)

plt.show()
```

```python
### Scatter plot
```

```python
fig, ax = plt.subplots() 

# encode a third variable by color
ax.scatter(df.Br, df.UK, c=df.index)


ax.set_xlabel("Production of apples (tonnes) in Brazil")
ax.set_ylabel("Production of apples (tonnes) in the UK")


plt.show()
```

```python
fig, ax = plt.subplots() 

#compare
ax.scatter(df.loc['1961':'1969'].Br, df.loc['1961':'1969'].UK, c='c', label='sixties')
ax.scatter(df.loc['1991':'1999'].Br, df.loc['1991':'1999'].UK, color='green', label='nineties')

ax.set_xlabel("Production of apples (tonnes) in Brazil")
ax.set_ylabel("Production of apples (tonnes) in the UK")

plt.legend()

plt.show()
```
