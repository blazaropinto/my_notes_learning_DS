Seaborn's default configurations that can be applied with `sns.set()`
These styles can override matplotlib and pandas plots as well.

```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read in the DataFrame
df = pd.read_csv('../datasets/clean/simpledf_applevalues.csv', index_col=0)
df.head()
```

```python
# Plot the pandas histogram
df['Br'].plot.hist()
plt.show()
#plt.clf()

# Set the default seaborn style
sns.set()

# Plot the pandas histogram again
df['Br'].plot.hist()
plt.show()
#plt.clf()
```

```python
#same data in two different styles in order to see how the styles change the output
sns.set_style('dark')
sns.distplot(df['Br'])
```

```python
sns.set_style('whitegrid')
sns.distplot(df['Br'])
```

```python
# Set the style to white
sns.set_style('white')

# Create a regression plot
sns.lmplot(x='Br',y='Pt', data=df)

# Remove the spines
sns.despine()
```

```python
# Set the style to white
sns.set_style('white')

# Create a regression plot
sns.lmplot(x='Br',y='Pt', data=df)

# Remove the spines
sns.despine(bottom=True, left=True)
```

```python
for style in ['white','dark','whitegrid','darkgrid','ticks']:
    sns.set_style(style)
    sns.distplot(df['UK'])
    plt.show()
```

* Seaborn supports assigning colors to plots using  matplotlib color codes

```python
# Set style, enable color code, and create a magenta distplot
sns.set(color_codes=True)
sns.distplot(df['Br'], color='m')
```

```python
sns.set_style('white')
sns.set_color_codes
sns.distplot(df['Br'], color='g')
```

```python
#  bright and colorblind palettes
sns.set_palette('bright')
sns.distplot(df['Br'])
```

```python
sns.set_palette('colorblind')
sns.distplot(df['Br'])
```

* Seaborn uses the  `set_palette()` function to define a palette
* `sns.palplot()` function displays a palette
* `sns.color_palette()` returns the current palette and is used for defining Custom Palettes

```python
for p in sns.palettes.SEABORN_PALETTES:
    print(p)
    sns.set_palette(p)
    sns.palplot(sns.color_palette())
    plt.show()
```

```python
#Create and display a Purples sequential palette containing 8 colors.
sns.palplot(sns.color_palette('Purples', 8))   # when the data has a consistent range from high to low
```

```python
sns.palplot(sns.color_palette("Paired", 12))   #when the data is not ordered
```

```python
sns.palplot(sns.color_palette( "BrBG", 12))   #when both the low and high values are interesting
```

```python
sns.palplot(sns.color_palette('husl',10))
```

```python
sns.palplot(sns.color_palette('coolwarm', 6))
```

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the distribution of data
sns.distplot(df['Br'], ax=ax)

# Create a more descriptive x axis label
ax.set(xlabel="Apple production in Brazil")
```

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the distribution of 1 bedroom rents
sns.distplot(df['Br'], ax=ax)

# Modify the properties of the plot
ax.set(xlabel="Apple production in Brazil",
       xlim=(0,2000000),
       title="Br")
```

```python
# Create a figure and axes. Then plot the data
fig, ax = plt.subplots()
sns.distplot(df['Br'], ax=ax)

# Customize the labels and limits
ax.set(xlabel="Apple production in Brazil",
       xlim=(0,2000000),
       title="Br")

# Add vertical lines for the median and mean
ax.axvline(x=df['Br'].median(), color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=df['Br'].mean(), color='b', label='Mean', linestyle='-', linewidth=2)

# Show the legend and plot the data
ax.legend()
```

```python
# Create a plot with 1 row and 2 columns that share the y axis label
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(7,4))

# Plot the distribution of 1 bedroom apartments on ax0
sns.distplot(df['Br'], ax=ax0)
ax0.set(xlabel="Brazil", xlim=(0,1500000))

# Plot the distribution of 2 bedroom apartments on ax1
sns.distplot(df['Pt'], ax=ax1)
ax1.set(xlabel="Portugal", xlim=(0,600000))
```
