```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set default size for figures
plt.rcParams['figure.figsize'] = (16, 5)
```

```python
# load data
simp_app = pd.read_csv('../datasets/clean/simpledf_applevalues.csv', index_col=0)
simp_app.head()
```

#### distplot
* similar to an histogram
* by default, generates a Gaussian Kernel Density Estimate (KDE)
* in order to plot a simple histogram, disable the `kde` and specify the number of `bins` to use
* alternatively, you can combine a `rug` plot and a kde curve while diactivating the histogram to view the distribution of data
* the plot can be further customized by passing arguments to the underlying function (dictionary to the parameter `kde_kws`, for instance)

```python
# '_' is the convention in python for variables not to be used again
_ = sns.distplot(simp_app['Br'])
plt.show()
```

```python
_ = sns.distplot(simp_app['Br'], kde=False, bins=14)
plt.show()
```

```python
_ = sns.distplot(simp_app['Br'], hist=False, rug=True, kde_kws={'shade':True})
plt.show()
```

#### regplot
* usage is similar to distplot
* generates a sca?er plot with a regression line
* The `data` and `x` and `y` variables must be always defined

```python
sns.regplot(x='Br', y='Pt', data=simp_app)
plt.show()
```

```python
sns.lmplot(x='Br', y='Pt', data=simp_app) #lmplot allows faceting (hue-organize by colors, col-organize data by columns...)
plt.show()
```
