# Types of graphs: categorical plots

**Categorical Data**
Data which takes on a limited and fixed number of values
Normally combined with numeric data

```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load data
pop = pd.read_csv('../datasets/clean/pop.csv', index_col=0)
pop.head()
```

```python
# load data
df = pd.read_csv('../datasets/clean/cropresidues_europe.csv', index_col=0, 
                 usecols=['Area', 'Item', 'Element', 'average_1961-2019', 'std_1961-2019'])
df.head()
```

### to show each observation: stripplot and swarmplot

```python
sns.stripplot(data=pop, y='Element', x='Value', jitter=True)  # jitter is useful when you have many points and they overlap
plt.show()
```

```python
sns.swarmplot(data=pop[:200], y='Element', x='Value')  #This plot uses a complex algorithm to place the observations in a manner where they do not overlap. 
                                # The downside of this approach is that the swarmplot() does not scale well to large datasets.
plt.show()
```

### for abstract representations of the data: boxplot, violinplot and boxenplot

```python
# the boxplot used to show several measures related to the distribution of data, 
# including the median, upper and lower quartiles, as well as outliers.
sns.boxplot(data=pop[:200], y='Element', x='Value')
plt.show()
```

```python
# The violinplot is a combination of a kernel density plot and a box plot 
# and can be suitable for providing an alternative view of the distribution of data. 
# Since the plot uses a kernel density calculation it does not show all data points.
# This can be useful for displaying large datasets but it can be computationally intensive to create.

sns.violinplot(data=pop[:100], y='Element', x='Value')
plt.show()
```

```python
# the boxenplot() can scale more effectively to large datasets than the boxplot() and violinplot() 
# The boxenplot() is a hybrid between a boxplot() and violinplot() and is relatively quick to render and easy to interpret.

sns.boxenplot(data=pop[:1000], y='Element', x='Value')
plt.show()
```

### to show statistical estimates: barplot, pointplot and countplot

```python
# barplot() shows an estimate of the value as well as a confidence interval.
sns.barplot(data=pop[:1000], y='Element', x='Value')
plt.show()
```

```python
sns.pointplot(data=pop[:1000], y='Element', x='Value')
plt.show()
```

```python
# countplot displays the number of instances of each variable
sns.countplot(data=df[:500], y='Element')
plt.show()
```
