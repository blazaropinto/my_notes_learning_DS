## Manipulating Lists

* to change values:subset lists, and then assign new elements to it using the equals sign

* If you use the plus sign with two lists, Python simply pastes together their contents in a single list.

Behind the scenes you're storing the 'address' of that list,where the list is in your computer memory
when you copy lists you copy the reference to the list, not the actual values themselves

* If you want to create a list y that points to a new list in the memory with the same values, you'll need to use something else than the equals sign.

```python
x = ["a", "b", "c"]
y = x
y[1] = "z"
y
```

```python
x
```

```python
x = ["a", "b", "c"]
y = list(x)
y = x[:] 
y[1] = "z"
y
```

```python
x
```
