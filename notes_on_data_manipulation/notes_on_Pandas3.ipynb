```python
#import libraries and simple dataframe
import pandas as pd
import matplotlib.pyplot as plt # to help with additional customization etc

pop = pd.read_csv('../datasets/clean/pop.csv', index_col=0, header=0)
display(pop.head())
```

```python
# Create a bar plot
pop.groupby('Element')['Value'].sum().plot(kind='bar')
```

```python
# Create a line plot
pop[(pop.Area == 'Germany') & (pop.Element == 'Total Population - Both sexes')].plot(x='Year', y='Value', kind='line')
```

```python
# Scatter plot
pop[(pop.Area == 'France') & (pop.Element == 'Rural population')].plot(x='Year', y='Value', kind='scatter', title='Rural population in France per year')
```

```python
# histogram
pop[(pop.Area == 'Greece') & (pop.Element == 'Total Population - Female')]['Value'].plot(kind='hist', bins=5, alpha=0.5)
pop[(pop.Area == 'Greece') & (pop.Element == 'Total Population - Male')]['Value'].plot(kind='hist', bins=5, alpha=0.5)
plt.legend(['Female', 'Male'])
plt.show()
```

```python
# load larger dataset with missing values
eu_apples_prod = pd.read_csv('../datasets/clean/apples_eu.csv', index_col=0, header=0)
display(eu_apples_prod.head())
```

```python
# check for missing values
display(eu_apples_prod.isna().any().head()) # any missing value in each column?
# visualize this info
plt.figure(figsize=(18,5))
eu_apples_prod.isna().sum().plot(kind='bar')
plt.title('Missing values for apple production per country')
plt.show()
```
