````python
#imports
import requests
from bs4 import BeautifulSoup
import pandas as pd
````

````python
#try for one day
url_0101 = 'https://www.bornglorious.com/world/birthday/?pf=901&pd=0101'
cont0101 = requests.get(url_0101).content
soup0101 = BeautifulSoup(cont0101, 'html.parser')

#get scientists boxes of info
boxes = soup0101.find_all('div', {'class':'col-sm-4'})

#get names for births on this date
for box in boxes[:5]:
    print(box.find('b').text)
    print(box.find('div',{'class':'col-xs-7'} ).find('b').text)
    try:
        print(box.find('div',{'class':'col-xs-7'} ).find_all('b')[1].text)
    except:
        print('place missing')
    print('\n')
````

````python
# try to loop over pages
url_gen = 'https://www.bornglorious.com/world/birthday/?pf=901&pd='
month = list(range(1,13))
day = list(range (1, 32))
exurl = url_gen + str(month[0]).zfill(2) + str(day[0]).zfill(2)
exurl
````

````python
#try for january
for d in day:
    
    #get url
    url = url_gen + str(month[0]).zfill(2) + str(d).zfill(2) 
    print(url)
    #get soup
    c = requests.get(url).content
    soup = BeautifulSoup (c, 'html.parser')
   
    #get scientists boxes of info
    boxes = soup.find_all('div', {'class':'col-sm-4'})

    #get names for births on this date
    for box in boxes[:5]:
        print(box.find('b').text)
        print(box.find('div',{'class':'col-xs-7'} ).find('b').text)
        try:
            print(box.find('div',{'class':'col-xs-7'} ).find_all('b')[1].text)
        except:
            print('place missing')
        print(box.find('div',{'class':'panel-footer'}).text)
        print('\n')
````

#### Loops
* the `break` statement provides you with the opportunity to exit out of a loop when an external condition is triggered.
* The `continue` statement gives you the option to skip over the part of a loop where an external condition is triggered, but to go on to complete the rest of the loop. That is, the current iteration of the loop will be disrupted, but the program will return to the top of the loop.
* The `pass` statement allows you to handle the condition without the loop being impacted in any way; all of the code will continue to be read unless a break or other statement occurs.

````python
# all data to dataframe
df = pd.DataFrame(columns=['month', 'day', 'birthdate', 'name', 'place', 'description'])
i = -1

for m in month:
    for d in day:
        #get url
        try:
            url = url_gen + str(m).zfill(2) + str(d).zfill(2) 
        except:
            continue
        #get soup
        c = requests.get(url).content
        soup = BeautifulSoup (c, 'html.parser')

        #get scientists boxes of info
        boxes = soup.find_all('div', {'class':'col-sm-4'})

        #get names for births on this date
        for box in boxes:
            i += 1
            df.loc[i, 'day'] = d
            df.loc[i, 'month'] = m
            df.loc[i, 'name'] = (box.find('b').text)
            try:
                df.loc[i, 'birthdate'] = (box.find('div',{'class':'col-xs-7'} ).find('b').text)
            except:
                print('unknown')
            try:
                df.loc[i, 'place'] = (box.find('div',{'class':'col-xs-7'} ).find_all('b')[1].text)
            except:
                df.loc[i, 'place'] = 'place missing'
            try:
                df.loc[i, 'description'] = (box.find('div',{'class':'panel-footer'}).text)
            except:
                df.loc[i, 'description'] = 'description missing'
            
df.head()
````
